#### Take subset based on condition 

```python
cond1 =  df1['mpg'] > 40 # relational operators --> output will always be boolean
cond1
```

```python
df1[cond1] # we are saying extract rows where value is True
```

```python
cond2 =  (df1['mpg'] > 40 ) & (df1['origin'] == 3)
df1[cond2]
```

```python
cond3 =  ~(df1['mpg'] > 40) # return the data where mpg value is less than 40
df1[cond3]
```

```python
# and --> comparing two values 
# & --> to compare two series 
```

```python
df1[ ~(df1['mpg'] > 40) ]# return the data where mpg value is less than 40
```
