```python
##Merge##
```

```python
df1 = pd.DataFrame({'id': [1,2,3,4,5], 'stat_marks': [94,95,99,98,91]})
df2 = pd.DataFrame({'id': [3,4,5,6,7], 'ml_marks': [92,90,94,89,95]})
df2
```

```python
df1 = pd.DataFrame({'id':[1,2,3,2,5], 'stat_marks': [94,95,9,49,96]})
df1
```

```python
df1.merge(df2, on= 'id',how = 'inner')
```

```python
df1.merge(df2, on= 'id',how = 'outer')
```

```python
df1.merge(df2, on= 'id',how = 'left')
```

```python
 df1.merge(df2, on= 'id',how = 'right')
```

### Concat 

```python
df1 = pd.DataFrame({'id': [1,2,3,4,5], 'stat_marks': [94,95,99,98,91]})
df2 = pd.DataFrame({'id': [3,4,5,6,7], 'ml_marks': [92,90,94,89,95]})
```

```python
pd.concat([df1,df2])
```
