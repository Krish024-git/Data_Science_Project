```python
plt.scatter(C,A, color = 'Red', marker = 'o')
plt.xlabel('Months')
plt.ylabel('Products sold')
plt.title('Sales Performance A')

plt.grid() # to add reference lines
plt.show() # to hide this <matplotlib.collections.PathCollection at 0x173ad5640>
```

```python
plt.scatter(C,A, color = 'Red', marker = 'o', label = 'Person A')
plt.scatter(C,B, color = 'Blue', marker = 'D', label = 'Person B')
plt.xlabel('Months')
plt.ylabel('Products sold')
plt.title('Sales Performance A')
plt.legend() # it displays value written inside label parameter
plt.grid() # to add reference lines
plt.show() 
```

```python
df1 = pd.read_csv('Auto.csv')
df1.head()
```

```python
# before data visualisation we need to clean the data ( missing values , outlier, duplicates , data type conversions.... )
```

```python
# Divide data into statistical data types 
# continuous : mpg, displacement, Horse Power, weight , acceleration 
# discrete: cylinders, year, origin, name
```

#### Bar Plot

```python
# bar plot can be created only for discrete data
# x, y needs to specifically mentioned 
```

```python
df1[df1['origin'] == 1]
```

```python
output = df1['origin'].value_counts()
output
```

```python
uniq = output.index
uniq
```

```python
freq = output.values
freq
```

```python
plt.bar(uniq, freq, color = ['Blue','Yellow','Red']) # colors will be assigned in order of uniq values
# uniq : 1,3,2 --> 1: BLUE, 3 --> yellow, 2--> Red
plt.xlabel('Months')
plt.ylabel('Products sold')
plt.title('Sales Performance B')
plt.grid() # to add reference lines
plt.show()
```

#### Pie chart

```python
# pie chart should be created for discrete data
```

```python
plt.pie(freq, labels = uniq)
plt.show()
```
