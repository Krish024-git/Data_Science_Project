```python
res = []
for i in df1['weight']:
    if i%2 == 0:
        res.append('Even')
    else:
        res.append('Odd')
df1['col1'] = res
df1.head()
```

```python
def evn_odd(i):
    if i%2 == 0:
        return 'Even'
    else:
        return 'Odd'
```

```python
evn_odd(2)
```

```python
df1['col2'] = df1['weight'].apply(evn_odd)
df1.head()
```

```python
# one time execution but dont want use loops

df1['col3'] =  df1['weight'].apply(lambda num: 'Even' if num%2 == 0 else 'Odd')
df1.head()
```
