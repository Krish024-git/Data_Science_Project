#### Array Mathematics

```python
ar1 = np.array([[10,20],[30,40]])
ar1
```

```python
np.sum(ar1)
```

```python
np.sum(ar1, axis = 0)
```

```python
np.sum(ar1, axis = 1)
```

```python
print(np.add(10,20))
print(np.subtract(10,20))
print(np.multiply(10,20))
print(np.divide(10,20))
```

```python
print(np.log(10))
print(np.sqrt(10))
print(np.exp(10))
print(np.sin(10))
```
