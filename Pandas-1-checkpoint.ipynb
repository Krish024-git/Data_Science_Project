#### Merge

```python
# SQL JOIN == Pandas Merge
```

```python
df1 = pd.DataFrame({'id': [1,2,3,4,5], 'stat_marks': [94,95,99,98,91]})
df2 = pd.DataFrame({'id': [3,4,5,6,7], 'ml_marks': [92,90,94,89,95]})
df2
```

```python
# to extract students data who appeared for both exams 
# select * from df1 join df2 on df1.id = df2.id

df1.merge(df2, on = 'id', how = 'inner')
```

```python
df1.merge(df2, on = 'id', how = 'left') # NaN --> SQL Null / empty
```

```python
df1.merge(df2, on = 'id', how = 'right')
```

```python
df1.merge(df2, on = 'id', how = 'outer')
```

#### Concat

```python
# SQL UNION == Pandas concat
```

```python
df1 = pd.DataFrame({'id': [1,2,3,4,5], 'stat_marks': [94,95,99,98,91]})
df2 = pd.DataFrame({'id': [6,7,8,9,10], 'stat_marks': [92,90,94,89,95]})

pd.concat([df1,df2])
```
