```python
# Array Slicing

import numpy as np

list1 = [10,20,30,40,50]
list1

# write a code to print 20,30,40 --> indexing
list1[1:4]

# indexing will always start from zero 
# end value is not included

ar1 = np.array([10,20,30,40,50])
ar1[1:4]

# 2D data structure

ar1 = np.arange(1,25)
ar1

ar2 = ar1.reshape(6,4)
ar2

ar1 = np.arange(1,25).reshape(6,4)
ar1

ar1[2:4,1:3]

ar1[2:6,0:4]

# if starting value is not mentioned --> it will be 0 by default

ar1[2:6,:4]

# if end value is not mentioned --> it extracts till last value available
ar1[2:, :4]

ar1[2:,:]

# if you want to extract from 2nd row and all the columns 
ar1[2:]

# to extract all the rows and 1,2,3 columns (index)
ar1[:, 1:4]

ar1[2,1]
ar1[2:3,1:2]


## Array manipulation
ar1 = np.array([[10,20],[30,40]])
ar2 = np.array([[50,60],[70,80]])

np.concatenate([ar1,ar2],axis = 0) # combine arrays at rows 
np.row_stack([ar1,ar2])
np.vstack([ar1,ar2])


np.concatenate([ar1,ar2],axis = 1) # combine arrays at columns
np.column_stack([ar1,ar2])
np.hstack([ar1,ar2])

ar3 = np.hstack([ar1,ar2,ar1])
ar3

np.hsplit(ar3,3) # split array into 3 equal portions
# to split array different levels use slicing

ar1 = np.array([[10,20],[30,40]])
ar2 = np.array([[50,60],[70,80]])
np.concatenate([ar1,ar2], axis=1)
np.concatenate([ar1,ar2], axis=0)

np.row_stack([ar1,ar2])
np.vstack([ar1,ar2])
```
